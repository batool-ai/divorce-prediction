```python
#----------------------------------------------
# üîπ 1. ÿßÿ≥ÿ™Ÿäÿ±ÿßÿØ ÿßŸÑŸÖŸÉÿ™ÿ®ÿßÿ™
#----------------------------------------------
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

#----------------------------------------------
# üîπ 2. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÖŸÜ ÿßŸÑŸÖŸÑŸÅ ÿßŸÑŸÖÿ≠ŸÑŸä
#----------------------------------------------
data = pd.read_csv(r"C:\Users\Lenovo\Downloads\divorce.csv", sep=';')

#----------------------------------------------
# üîπ 3. ŸÅÿµŸÑ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ (X) ÿπŸÜ ÿßŸÑÿπŸÖŸàÿØ ÿßŸÑŸáÿØŸÅ (y)
#----------------------------------------------
X = data.drop('Class', axis=1)
y = data['Class']

#----------------------------------------------
# üîπ 4. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ•ŸÑŸâ ÿ™ÿØÿ±Ÿäÿ® ŸàÿßÿÆÿ™ÿ®ÿßÿ±
#----------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

#----------------------------------------------
# üîπ 5. ÿ™Ÿàÿ≠ŸäÿØ ÿßŸÑŸÇŸäŸÖ (Normalization)
#----------------------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#----------------------------------------------
# üîπ 6. ÿ®ŸÜÿßÿ° ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ (Random Forest)
#----------------------------------------------
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

#----------------------------------------------
# üîπ 7. ÿßŸÑÿ™ŸÜÿ®ÿ§ ŸàÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
#----------------------------------------------
y_pred = model.predict(X_test_scaled)

print("üîπ ÿßŸÑÿØŸÇÿ© ÿßŸÑŸÉŸÑŸäÿ© (Accuracy):", accuracy_score(y_test, y_pred))
print("\nüîπ ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ™ÿµŸÜŸäŸÅ:")
print(classification_report(y_test, y_pred))

#----------------------------------------------
# üîπ 8. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßÿ±ÿ™ÿ®ÿßŸÉ (Confusion Matrix)
#----------------------------------------------
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

#----------------------------------------------
# üîπ 9. ÿ£ŸáŸÖ ÿßŸÑÿ£ÿ≥ÿ¶ŸÑÿ© ÿßŸÑÿ™Ÿä ÿ™ÿ§ÿ´ÿ± ÿπŸÑŸâ ÿßŸÑÿ∑ŸÑÿßŸÇ
#----------------------------------------------
importances = pd.Series(model.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(10)
print("\nüîπ ÿ£ŸáŸÖ 10 ÿ£ÿ≥ÿ¶ŸÑÿ© ÿ™ÿ§ÿ´ÿ± ŸÅŸä ÿßŸÑŸÜÿ™Ÿäÿ¨ÿ©:\n", top_features)

plt.figure(figsize=(8,5))
sns.barplot(x=top_features.values, y=top_features.index, palette="coolwarm")
plt.title("Top 10 Questions Influencing Divorce Prediction")
plt.xlabel("Importance")
plt.show()
```
